```python
import numpy as np
```

```python
g=9.8
L=1
x = np.linspace(0, 4*np.pi, 30) #values of angular position
y = np.linspace(-(0.3)*(g/L)*np.pi, (0.3)*(g/L)*np.pi, 30) #values of angular speed
uu, vv = np.meshgrid(x, y) ##uu is x repeated 30times,vv is each value of y repeated 30 times
```
